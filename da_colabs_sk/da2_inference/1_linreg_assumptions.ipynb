## Štatistická inferencia pomocou lineárnej regresie

### Overenie predpokladov

Ak sa používa lineárna regresia na štatistickú inferenciu, okrem koeficientov lineárneho modelu sa vypočítajú aj určité štatistické ukazovatele (napr. p-hodnoty, BIC, ...). Výpočet týchto ukazovateľov vychádza z určitých predpokladov a tieto musia platiť, ak majú výsledky byť validné. Preto pred vyvodením akýchkoľvek záverov je potrebné overiť, či riešený problém tieto predpoklady spĺňa.

Konkrétne sa predpokladá [[islr](#islr),[duke](#duke)]:

# **Linearita:**  medzi prediktormi a predpovedanou premennou existuje lineárna závislosť;


# **Bez multicolinearity:**  prediktory nie sú vzájomne príliš silne korelované;


# **Nezávislosť rezíduí:**  rezíduá (chyby predikcie) na jednotlivých vzorkách sú vzájomne štatisticky nezávislé (nie je medzi nimi žiadna skrytá závislosť, napr. medzi rezíduami za sebou nasledujúcich hodnôt v časovom rade a pod.);


# **Homoskedasticita rezíduí:**  rozptyl rezíduí je konštantný, t.j. nezávisí od žiadneho z prediktorov, od času a pod.;


# **Normalita rezíduí:**  rezíduá sú normálne rozdelené a stredná hodnota je nulová;

#### Ošetrenie nelineárnych vzťahov

Existuje niekoľko spôsobov, ako ošetriť silno nelineárne vzťahy medzi premennými – výber vhodného prístupu závisí typicky od toho, prečo je vzťah nelineárny. Môžeme uviesť povedzme nasledovné možnosti [[duke](#duke)]:

* **Aplikácia nelineárnej transformácie:**  Na prediktor aplikujeme nelineárnu transformáciu (napr. logaritmus, mocninu, exponenciálnu funkciu, ...). Typicky je potrebné v tomto postupovať opatrne – nie je v poriadku aplikovať ľubovoľnú nelineárnu transformáciu len preto, aby tam nejaká bola – výber transformácie musí dávať v kontexte konkrétnej závislosti zmysel. Pozorované dáta sa dajú často vcelku dobre aproximovať napr. pomocou polynómu niektorého vyššieho stupňa, avšak takéto polynomické modely majú často problém zovšeobecňovať a sú obvykle veľmi zlé pri extrapolácii (predikcii mimo pozorovaného rozsahu dát).
